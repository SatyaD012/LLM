```python


import torch
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification, AutoModelForCausalLM
from huggingface_hub import notebook_login
import gradio as gr

```

```python
# Hugging Face login - it will prompt you to enter your token
notebook_login()
```

```python
# Initialize device (use GPU if available)
device = torch.device('cuda' if torch.cuda.is_available() else torch.device('cpu'))
```

```python
# Load sentiment analysis model and tokenizer
sentiment_model_name = "tabularisai/robust-sentiment-analysis"
sentiment_tokenizer = AutoTokenizer.from_pretrained(sentiment_model_name)
sentiment_model = AutoModelForSequenceClassification.from_pretrained(sentiment_model_name)
```

```python
# Setup sentiment analysis pipeline
sentiment_pipeline = pipeline(
    "sentiment-analysis",
    model=sentiment_model,
    tokenizer=sentiment_tokenizer,
    max_length=512,  # Adjust max_length to a more reasonable size for short inputs
)
```

```python
# Mapping sentiment labels
sentiment_labels = {
    'LABEL_0': 'Very Negative',
    'LABEL_1': 'Negative',
    'LABEL_2': 'Neutral',
    'LABEL_3': 'Positive',
    'LABEL_4': 'Very Positive'
}
```

```python
# Define sentiment-based prompts for chatbot responses
sentiment_prompts = {
    'Very Positive': 'We truly appreciate your positive feedback and will keep striving for excellence! Based on your input: ',
    'Positive': 'Thank you for your kind words! We are delighted to hear your feedback. Based on your input: ',
    'Neutral': 'Could you provide more details? We’d love to understand how we can improve. Based on your input: ',
    'Negative': 'We apologize for your experience. Please let us know how we can make it better. Based on your input: ',
    'Very Negative': 'We are extremely sorry to hear about your experience. Your feedback is important to us, and we are taking it very seriously. Based on your input: '
}
```

```python
# Load chatbot model and tokenizer (Llama 2)
llm_model_name = "NousResearch/Llama-2-7b-chat-hf"
chat_tokenizer = AutoTokenizer.from_pretrained(llm_model_name)
chat_model = AutoModelForCausalLM.from_pretrained(llm_model_name, torch_dtype=torch.float16, device_map="auto")
chat_tokenizer.use_default_system_prompt = False

chat_pipeline = pipeline(
    "text-generation",
    model = chat_model,
    tokenizer = chat_tokenizer,
    torch_dtype=torch.float16,
    device_map="auto",
    max_length=2048,
    truncation=True,
)
```

```python

# Function to determine sentiment and generate response prompt
def analyze_sentiment(text):
    result = sentiment_pipeline(text)
    label = result[0]['label']
    sentiment = sentiment_labels[label]
    prompt = sentiment_prompts[sentiment] + f'"{text}"'
    return sentiment, prompt
```

```python
# Function to format conversation history
def format_history(history):
    formatted_history = ["You are a customer support agent. Respond appropriately based on customer sentiment.\n\n"]
    for customer, bot in history:
        formatted_history.append(f"Customer: {customer}\nBot: {bot}\n")
    return "\n".join(formatted_history)
```

```python
# Main chatbot function that combines sentiment analysis and response generation
def chatbot_response(user_input, chat_history):
    # Get sentiment and response prompt
    sentiment, prompt = analyze_sentiment(user_input)

    # Format conversation history
    conversation = format_history(chat_history)

    # Create a cleaner, structured prompt for the bot
    structured_prompt = f"{conversation}\nUser Input: {user_input}"

    # Generate response using LLM
    response = chat_pipeline(
        structured_prompt,
        max_new_tokens=150,
        return_full_text=False
    )[0]['generated_text']

    # Clean up response text
    return response.split("Response:")[-1].strip(), sentiment
```

```python

# Gradio interface function
def bot_interface(input, history):
    response, emotion = chatbot_response(input, history)
    return f"Customer's Emotion: {emotion}\n\nAssistant: {response.split('Assistant:')[-1]}"

# Set up Gradio Chat Interface
frontend = gr.ChatInterface(
    fn=bot_interface,
    title="Customer Support Chatbot for a 5⭐ Restaurant",
    description="This chatbot provides tailored responses based on customer sentiment.",
    examples=[
        "I'm really frustrated with your service. Your reception was disappointing.",
        "Your food is amazing! I'm so happy I found your restaurant.",
        "The food was ok. Nothing special."
    ],
)

frontend.launch(share=True)
```
